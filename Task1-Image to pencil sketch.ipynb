In this task , we need to read the image in RBG format and then convert it to a grayscale image.This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image.

```python
#imprting rquired lybraries
import cv2
from matplotlib import pyplot as plt
```

### Reading Image in RGB Formate
##### image is in BGR format need to convert in RGB format using method cv2.cvtColor()

```python
image=cv2.imread('myimage.jpeg')
O_image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10,10))
plt.imshow(O_image)
plt.title("Image In RGB Format")
plt.axis('off');
```

### Converting Image To Grayscale Image 

```python
G_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(10,10))
plt.imshow(G_image,cmap='gray')
plt.title("Grayscale Image")
plt.axis('off');
```

### Converting Grayscale Image in Inverted Grayscale Image OR Negative Image

```python
N_g_image=cv2.bitwise_not(G_image)
plt.figure(figsize=(10,10))
plt.imshow(N_g_image,cmap='gray')
plt.title("Inverted Grayscale OR Negative Image")
plt.axis('off');
```

### Converting Grayscale Image In Inverted Blurry Image 

```python
B_image=cv2.GaussianBlur(N_g_image,(61,61),0)
plt.figure(figsize=(10,10))
plt.imshow(B_image,cmap='gray')
plt.title("Blure Image")
plt.axis('off');
```

### Converting Grayscale Image in Pencile Sketch

```python
S_image=cv2.divide(G_image,100-N_g_image,scale=450)
plt.figure(figsize=(10,10))
plt.imshow(S_image,cmap='gray')
plt.title("Pencil Sketch Image")
plt.axis('off');
```
